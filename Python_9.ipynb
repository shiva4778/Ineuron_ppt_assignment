```python
# Get all the cars that have an electric range of more than 100, and were made after 2015
df['Model Year']=pd.to_datetime(df['Model Year'],format='%Y')
df[(df['Model Year'].dt.year>2015) & (df['Electric Range']>100)]
```

```python
#Creating an sepearte column for latitude
def l1(i):
    
    k=i.split(' ')[1][1:]
    return k
    
df['longitude']=df['Vehicle Location'].apply(l1)
```

```python
# Creting an seperate colmn for longitude

def l2(i):
    
    k=i.split(' ')[2][:-1]
    return k
    
df['latitude']=df['Vehicle Location'].apply(l2)
```

```python
df['longitude']=df['longitude'].astype(float)
df['latitude']=df['latitude'].astype(float)
```
